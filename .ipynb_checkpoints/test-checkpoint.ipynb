```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df_test=pd.read_csv('./Test_set.csv')
df_test
```

```python
df_test.info()
```

```python
df_test.describe()
```

```python
df_test.dropna(inplace=True)
df_test.shape
```

```python
df_test.isnull().sum()
```

```python
df_test['Date_of_Journey'].value_counts()
```

```python
df_test['Journey_Day']=pd.to_datetime(df_test['Date_of_Journey'],format='%d/%m/%Y').dt.day
df_test['Journey_Month']=pd.to_datetime(df_test['Date_of_Journey'],format='%d/%m/%Y').dt.month
df_test
```

```python
df_test['Journey_Weekday']=pd.to_datetime(df_test['Date_of_Journey'],format='%d/%m/%Y').dt.day_name()
df_test.drop(columns=['Date_of_Journey'],inplace=True)
df_test.shape
```

```python

journey_day=df_test[['Journey_Weekday']]
journey_day_ohe=pd.get_dummies(journey_day,drop_first=True)
journey_day_ohe.shape[1]
```

```python
df_test['Dep_Time'].value_counts()
df_test['Dep_Hour']=pd.to_datetime(df_test['Dep_Time']).dt.hour
df_test['Dep_Min']=pd.to_datetime(df_test['Dep_Time']).dt.minute
df_test
```

```python
df_test.drop(columns=['Dep_Time'],inplace=True)
df_test.shape
```

```python
df_test['Arrival_Time'].value_counts()
df_test['Arr_Hour']=pd.to_datetime(df_test['Arrival_Time']).dt.hour
df_test['Arr_Min']=pd.to_datetime(df_test['Arrival_Time']).dt.minute
df_test
```

```python
df_test.drop(columns=['Arrival_Time'],inplace=True)
df_test.shape
```

```python
duration = list(df_test["Duration"])

for i in range(len(duration)):
    if len(duration[i].split()) != 2:    # Check if duration contains only hour or mins
        if "h" in duration[i]:
            duration[i] = duration[i].strip() + " 0m"   # Adds 0 minute
        else:
            duration[i] = "0h " + duration[i]           # Adds 0 hour

duration_hours = []
duration_mins = []
for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep = "h")[0]))    # Extract hours from duration
    duration_mins.append(int(duration[i].split(sep = "m")[0].split()[-1]))   # Extracts only minutes from duration
df_test["Duration_hours"] = duration_hours
df_test["Duration_mins"] = duration_mins
df_test.drop(["Duration"], axis = 1, inplace = True)
```

```python
df_test['Airline'].value_counts()
```

```python
sb.countplot(df_test,y='Airline')
```

```python

airlines=df_test[['Airline']]
airline_ohe=pd.get_dummies(airlines,drop_first=True)
airline_ohe.shape[1]
```

```python
df_test['Source'].value_counts()
```

```python
source=df_test[['Source']]
source_ohe=pd.get_dummies(source,drop_first=True)
source_ohe.shape[1]
```

```python

df_test['Destination'].value_counts()
```

```python
dest=df_test[['Destination']]
dest_ohe=pd.get_dummies(dest,drop_first=True)
dest_ohe.shape[1]
```

```python
df_test.drop(columns=['Additional_Info','Route'],inplace=True)
df_test
```

```python
df_test['Total_Stops'].value_counts()
```

```python
df_test.replace({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4},inplace=True)
df_test_new = pd.concat([df_test,source_ohe,dest_ohe,airline_ohe,journey_day_ohe],axis=1)
df_test_new.drop(columns=['Source','Destination','Airline','Journey_Weekday'],axis=1,inplace=True)
df_test_new
```
